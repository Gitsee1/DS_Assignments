●	Handling missing values appropriately (e.g. imputation deletion etc)

```python
#Since there is no categorical data we shall replace the nan values by the mean 
#Mean Imputation of all the cols
meanAC = data_cleaned1['AC'].mean()
print ('The mean value of col AC is: ',meanAC)
meanFM = data_cleaned1['FM'].mean()
print ('The mean value of col FM is: ',meanFM)
meanUC = data_cleaned1['UC'].mean()
print ('The mean value of col UC is: ',meanUC)
meanDL = data_cleaned1['DL'].mean()
print ('The mean value of col DL is: ',meanDL)
meanDS = data_cleaned1['DS'].mean()
print ('The mean value of col DS is: ',meanDS)
meanDP = data_cleaned1['DP'].mean()
print ('The mean value of col DP is: ',meanDP)
meanASTV = data_cleaned1['ASTV'].mean()
print ('The mean value of col ASTV is: ',meanASTV)
meanMSTV = data_cleaned1['MSTV'].mean()
print ('The mean value of col MSTV is: ',meanMSTV)
meanALTV = data_cleaned1['ALTV'].mean()
print ('The mean value of col ALTV is: ',meanALTV)
meanMLTV = data_cleaned1['MLTV'].mean()
print ('The mean value of col MLTV is: ',meanMLTV)
meanWidth = data_cleaned1['Width'].mean()
print ('The mean value of col Width is: ',meanWidth)
meanTend = data_cleaned1['Tendency'].mean()
print ('The mean value of col Tendency is: ',meanTend)
meanNSP = data_cleaned1['NSP'].mean()
print ('The mean value of col NSP is: ',meanNSP)
```

```python
data_cleaned1['AC'] = data_cleaned1['AC'].fillna(meanAC)
data_cleaned1['FM'] = data_cleaned1['FM'].fillna(meanAC)
data_cleaned1['UC'] = data_cleaned1['UC'].fillna(meanAC)
data_cleaned1['DL'] = data_cleaned1['DL'].fillna(meanAC)
data_cleaned1['DS'] = data_cleaned1['DS'].fillna(meanAC)
data_cleaned1['DP'] = data_cleaned1['DP'].fillna(meanAC)
data_cleaned1['ASTV'] = data_cleaned1['ASTV'].fillna(meanAC)
data_cleaned1['MSTV'] = data_cleaned1['MSTV'].fillna(meanAC)
data_cleaned1['ALTV'] = data_cleaned1['ALTV'].fillna(meanAC)
data_cleaned1['MLTV'] = data_cleaned1['MLTV'].fillna(meanAC)
data_cleaned1['Width'] = data_cleaned1['Width'].fillna(meanAC)
data_cleaned1['Tendency'] = data_cleaned1['Tendency'].fillna(meanAC)
data_cleaned1['NSP'] = data_cleaned1['NSP'].fillna(meanAC)
```
